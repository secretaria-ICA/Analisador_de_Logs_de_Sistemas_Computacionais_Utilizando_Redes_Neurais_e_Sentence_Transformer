#Montando a Rede Neural

```python
#Utilizar uma versão do tensorflow anterior a 2.0
%tensorflow_version 1.x
```

```python
#NumberOfClasses = len(y_train.unique())
NumberOfClasses = len(data['class'].unique())

from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
y_train = labelencoder_y.fit_transform(y_train)
y_test = labelencoder_y.fit_transform(y_test)
print(y_train)
```

## Criação  e dimensionamento

```python
from keras import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```python
X_train.shape[1:]
```

```python
RN = Sequential()
# activation = sigmoid / relu / 
RN.add(Dense(384, input_shape = X_train.shape[1:], activation = 'relu'))
RN.add(Dense(256,activation = 'relu'))
# Função de ativação obrigatoriamente sigmóide pelo fato da saída ser binária
RN.add(Dense(NumberOfClasses/2,activation = 'sigmoid'))

RN.summary()
```

```python
# treinamento
##from keras.utils import to_categorical
np.random.seed(0)
#### ? tf.set_random_seed=0
#optimizer: adam/sgd
#RN.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy']) 
RN.compile(optimizer = 'sgd', loss = 'binary_crossentropy', metrics = ['accuracy']) 
trainedRN = RN.fit(X_train,y_train, epochs = 200, verbose = 1)
```

## Avaliação da rede neural

```python
# avaliação
score = trainedRN.model.evaluate(X_test, y_test, verbose = 1)
print('Test score:', score[0])
print('Test mape:', score[1], '%')
```
